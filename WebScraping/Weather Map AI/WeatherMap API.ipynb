# Data Acquisition
- Application Programming Interface-Set of Routines running on a web server
- Connects Application and Database
- Specify Data with URL
- Retrive Data and Parse this
- API may return data in JSON,XML data formats

#### APIs allow you to access data from a webserver.It is a software intermediary that allows two applications to talk to each other.Each time you use an app like Facebook, send an instant message, or check the weather on your phone, you're using an API.

#### In python we have two libraries which helps us interact with APIs
- urllib library
- Request Library

# 1. Weather Map API

#### We will first make a GET request using the urllib library

```python
from urllib.request import urlopen
```

```python
api_url="http://api.openweathermap.org/data/2.5/weather?q=London,uk&APPID=9d4f5f3041f4cd4efe192635a45e74e6"
print(type(api_url))
```

#### We have got a HTTPResponse, now we ll have to parse the result using read()

```python
url_result=urlopen(api_url)
print(url_result)
```

```python
data=url_result.read()
print(type(data))
```

#### The data is in the form of bytes, we need to convert it into JSON

#### Json is a python like dictionary

```python
print(data)
```

#### loads() method can be used to parse a valid JSON string and convert it into a Python Dictionary. It is mainly used for deserializing native string, byte, or byte array which consists of JSON data into Python Dictionary

```python
import json
```

```python
json_data=json.loads(data)
```

```python
print(json_data)
```

```python
json_data['coord']
```

```python
json_data['weather']
```

```python
json_data['name']
```

```python
json_string=json.dumps(json_data)
print(json_string)
```

```python
print(type(json_data))
print(type(json_string))
```

# 3. Facebook API

## Requests Library
- Most Popular Python Library
- Wrapper over urllib
- Works well with Python3

```python
import requests
```

```python
url = "http://graph.facebook.com/{user_id}/picture?type=large"
```

```python
r = requests.get(url)
```

```python
print(type(r))
```

```python
print(r)
```

```python
print(r.content)
```

```python
with open("sample_pic.jpg",'wb') as f:
    f.write(r.content)
```
